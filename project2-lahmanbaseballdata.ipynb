```python
data=pd.read_csv("Batting.csv")
```

```python
data
```

```python
data['yearID']=data['yearID'].astype('int')
data['3B']=data['3B'].astype('int')
data.head()
```

```python
years=data['yearID']
triples=data['3B']
plt.plot(years,triples)
plt.show()
        
```

```python
plt.plot(years,triples)
plt.xlabel('YEARS')
plt.ylabel('TRIPLES')
plt.title('HISTROGRAM 3B')
plt.show()
```

scatter plot relating triples(3B)and steals(SB)

```python

triples=data['3B']
steals=data['SB']
plt.plot(triples,steals)
plt.show()
```

```python
plt.scatter(triples,steals)
plt.xlabel('TRIPLES')
plt.ylabel('STEALS')
plt.title('SCATTER PLOT 3B and SB')
plt.show()
```
